```python
#EXPLORATORY DATA ANALYSIS
#UPLOAD THE DATASET AND CREATE DATAFRAME
#DATAFRAME - TABLED DATA , SET - RAW ,UNPROCESSED DATA
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Mukunthan14/RINEX-4-MAY-PROJECT-2/main/ds_salaries.csv')
df
```

```python
df.info
```

```python
df.nunique()
```

```python
#DATA CLEANING
df = df.drop(['Unnamed: 0'], axis=1)
df
```

```python
#VISUALISATION - SEABORN
import seaborn as sns
import matplotlib.pyplot as plt
#1)WHAT IS THE DISTRIBUTION OF JOBS/ HOW MANY PEOPLE IN EACH JOB?
plt.figure(figsize=(20,7))
df['job_title'].value_counts().plot(kind = 'bar')
#THE MOST POPULAR JOBS ARE DATA SCIENTIST, DATA ENGINEER AND DATA ANALYST, LEAST POPULAR IS DATA ANALYTICS LEAD
#DATA SCIENCE AND MACHINE LEARNING JOBS ARE VERY POPULAR, WHILE FINANCE RELATED AND MANAGERIAL POSITIONS ARE LESS POPULAR
```

```python
#2)HOW MANY OF THEM ARE PART TIME/FULL TIME, ETC....
df['employment_type'].value_counts().plot(kind = 'bar')
#MOST OF THEM ARE FULL TIME, SOME PART TIME, VERY LITTLE AMOUNT ARE CONTRACT/FREE LANCE
```

```python
#3)WHICH CURRENCY IS MOST USED?
df['salary_currency'].value_counts().plot(kind = 'bar')
#USD IS MOST USED, NEXT TO EUROS, GBP AND INR. THIS COULD BE BECAUSE MOST OF THE WORKERS ARE FROM US
```

```python
#4)WHAT IS THE DISTRIBUTION OF EMPLOYEES ACROSS THE WORLD?
plt.figure(figsize=(20,7))
df['employee_residence'].value_counts().plot(kind = 'bar')
#AS STATED ABOVE, THIS DISTRIBUTION EXPLAINS THE CURRENCY DISTRIBUTION
```

```python
#5)DISTRIBUTION OF REMOTE WORKERS
df['remote_ratio'].value_counts().plot(kind = 'bar')
#THUS MOST OF THEM ARE FULLY REMOTE WORKERS(WORK FROM HOME), SOME ARE PARTIALLY REMOTE AND SOME ARE NOT REMOTE
```

```python
#6)EXPERIENCE LEVEL OF EMPLOYEES
df['experience_level'].value_counts().plot(kind = 'bar')
#MOST OF THEM ARE SENIOR LEVEL AND JUNIOR/INTERMEDIATE LEVEL, VERY FEW ARE EXPERT LEVEL
```

```python
#7)CORRELATION BTW EXPERIENCE LEVEL AND SALARY
sns.barplot(x=df['experience_level'], y=df['salary_in_usd'])
#CLEARLY MORE EXPERIENCE LEVEL EQUALS HIGHER SALARY 
```

```python
#8)CORRELATION BTW JOB TITLE AND SALARY
plt.figure(figsize=(20,7))
sns.barplot(x=df['job_title'], y=df['salary_in_usd'])
plt.xticks(rotation=90)
#DATA SCIENTIST HAS AVERAGE SALARY OF 100000, PRINCIPAL DATA ENGINEER AND FINANCIAL DATA ANALYST HAVE HIGHEST SALARIES, PRODUCT DATA ANALYST AND BIG DATA ENGINEER HAVE BELOW AVEAGE SALARY
```

```python
#9)CORRELATION BTW YEAR THE SALARY WAS PAID AND SALARY
sns.barplot(x=df['work_year'], y=df['salary_in_usd'])
#AVAERAGE SALARY HAS INCREASED IN 2022
```

```python
#10)CORRELATION BTW COMPANY SIZE AND SALARY
sns.barplot(x=df['company_size'], y=df['salary_in_usd'])
#CLEARLY LARGER COMPANIES PAY MORE SALARY
```

```python
#11)CORRELATION BTW EMPLOYEE RESIDENCE AND SALARY
plt.figure(figsize=(20,7))
sns.barplot(x=df['employee_residence'], y=df['salary_in_usd'])
#US HAS HIGH AVERAGE SALARY. RUSSIA,JAPAN,AUS HAVE LESS AVERAGE SALARY BUT POTENTIAL FOR HIGHER SALARY. INDIA HAS BELOW AVERAGE SALARY
```

```python
#12)CORRELATION BTW REMOTE RATIO AND SALARY
sns.barplot(x=df['remote_ratio'], y=df['salary_in_usd'])
#FULLY REMOTE AND NOT REMOTE EMPLOYEES HAVE MORE SALARY, WHILE PARTIALLY REMOTE EMPLOYEES HAVE LEAST SALARY
```

```python
#13)CORRELATION BTW COMPANY LOCATION AND SALARY
plt.figure(figsize=(20,7))
sns.barplot(x=df['company_location'], y=df['salary_in_usd'])
#SIMILAR TO EMPLOYEE RESIDENCE
```

```python
#14)CORRELATION BTW JOB TITLE, EXPERIENCE LEVEL AND SALARY
plt.figure(figsize=(20,7))
plt.xticks(rotation=90)
sns.scatterplot(x=df['job_title'], y=df['salary_in_usd'], hue=df['experience_level'], data=df)
#MOSTLY SENIOR LEVEL EMPLOYEES HAVE HIGHER SALARIES, SOMETIMES EVEN MORE THAN EXPERTS IN OTHER JOBS
```

```python
#15)CORRELATION BTW JOB TITLE, COMPANY SIZE AND SALARY
plt.figure(figsize=(20,7))
plt.xticks(rotation=90)
sns.scatterplot(x=df['job_title'], y=df['salary_in_usd'], hue=df['company_size'], data=df)
#LARGE COMPANIES DO NOT ALWAYS OFFER HIGHER SALARY. MEDIUM SIZE COMPANIES OFFER ON AVERAGE HIGHER SALARIES, SOMETIMES EVEN MORE THAN SOME LARGE COMPANIES
```

```python
#16)CORRELATION BTW JOB TITLE, COMPANY LOCATION AND SALARY
plt.figure(figsize=(20,7))
plt.xticks(rotation=90)
sns.scatterplot(x=df['job_title'], y=df['salary_in_usd'], hue=df['company_location'], data=df)
#EVEN IN THE SAME JOB, AMERICAN COMPANIES PAY HIGHER SALARIES THAN OTHER COUNTRY'S COMPANIES
```
